```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Loadin the data
df = pd.read_csv('AirPassengers.csv', parse_dates=['Month'])
df.set_index('Month', inplace=True)
```

```python
# Create lag features
def create_lag_features(data, lags):
    for lag in lags:
        data[f'lag_{lag}'] = data['#Passengers'].shift(lag)
    return data

df = create_lag_features(df, lags=[1, 2, 3, 4, 5])
df.dropna(inplace=True)
```

```python
# Split the data
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size], df.iloc[train_size:]

# Prepare the training data
X_train = train.drop(['#Passengers'], axis=1)
y_train = train['#Passengers']

# Prepare the testing data
X_test = test.drop(['#Passengers'], axis=1)
y_test = test['#Passengers']
```

```python
# Convert the data to DMatrix format
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# Set up the parameters for the XGBoost model
params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
}
```

```python
# Train the model
model = xgb.train(params, dtrain, num_boost_round=100, evals=[(dtest, 'test')], early_stopping_rounds=10)
```

```python
# Make predictions
predictions = model.predict(dtest)
```

```python
# Convert predictions to a DataFrame
pred_df = pd.DataFrame({'actual': y_test, 'predicted': predictions})
print(pred_df.head())
```

```python
# Calculating the RMSE  root mean squared error 
rmse = mean_squared_error(y_test, predictions, squared=False)
print(f'RMSE: {rmse}')
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, predictions, label='Predicted')
plt.title('XGBoost Time Series Forecasting')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```
